-----------Converting Various Extensions to jpg------------- 

For Contaminated

```python
path="/content/drive/MyDrive/Cont"
for x in os.listdir(path):
  z=os.path.join(path,x)
  for m in os.listdir(z):
    if(m.endswith(".jpeg")):                           ##jpeg,jpg,png
      newp=os.path.join(z,m)
      img=cv.imread(newp)
    
      # Getting the image name without extension
      imgname=m[:-4]
      newpath="/content/drive/MyDrive/contaminated1/"+x
      # setting the path for saving the jpg image
      path_to_save=newpath+'/'+imgname+'.jpg'
        
      # Saving the image at same location where txt file is stored using opencv in jpg format
      cv.imwrite(path_to_save,img)
```

For Edible

```python
path="/content/drive/MyDrive/Edib"
for x in os.listdir(path):
  z=os.path.join(path,x)
  for m in os.listdir(z):
    if(m.endswith(".jpg")):                           ##jpeg,jpg,png
      newp=os.path.join(z,m)
      img=cv.imread(newp)
    
      # Getting the image name without extension
      imgname=m[:-4]
      newpath="/content/drive/MyDrive/edible1/"+x
      # setting the path for saving the jpg image
      path_to_save=newpath+'/'+imgname+'.jpg'
        
      # Saving the image at same location where txt file is stored using opencv in jpg format
      cv.imwrite(path_to_save,img)
```

---------Indexing of Images-----------

For Contaminated

```python
i=0
path="/content/drive/MyDrive/contaminated1"
for filename in os.listdir(path):
  z=os.path.join(path,filename)
  for x in os.listdir(z):
    dest =str(i+1)+".jpg"
    source=os.path.join(z,x)
    dest=os.path.join(z,dest) 
    os.rename(source,dest)
    i+=1
```

For Edible

```python
i=0
path="/content/drive/MyDrive/edible1"
for filename in os.listdir(path):
  z=os.path.join(path,filename)
  for x in os.listdir(z):
    dest =str(i+1)+".jpg"
    source=os.path.join(z,x)
    dest=os.path.join(z,dest) 
    os.rename(source,dest)
    i+=1
```

-------Resize of images (target size is 200*200)--------

For Contaminated

```python
path="/content/drive/MyDrive/contaminated1"
for i in (os.listdir(path)):
  new_path=os.path.join(path,i)
  for j in os.listdir(new_path):
    y="/content/drive/MyDrive/train_data/train_contaminated"    ##destination folder
    x=os.path.join(y,j)
    z=os.path.join(new_path,j)
    img=cv.imread(z)
    newimg=cv.resize(img,(200,200),cv.INTER_AREA)
    cv.imwrite(x,newimg)
```

For Edible

```python
path="/content/drive/MyDrive/edible1"
for i in (os.listdir(path)):
  new_path=os.path.join(path,i)
  for j in os.listdir(new_path):
    y="/content/drive/MyDrive/train_data/train_edible"    ##destination folder
    x=os.path.join(y,j)
    z=os.path.join(new_path,j)
    img=cv.imread(z)
    newimg=cv.resize(img,(200,200),cv.INTER_AREA)
    cv.imwrite(x,newimg)
```
